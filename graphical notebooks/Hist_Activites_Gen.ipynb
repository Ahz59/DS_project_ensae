```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_m = pd.read_csv('data/speeddating modified.csv')
df_p = pd.read_csv('data/speeddating personal.csv')
```

```python
activités = ['gender', 'sports', 'tvsports', 'exercise', 'dining', 'museums',
             'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater', 'movies',
             'concerts', 'music', 'shopping', 'yoga']
```

```python
df_p[df_p['gender'] == 'male'][activités].mean(numeric_only=True)
```

```python
df_activ = df_p[activités].groupby('gender').mean(numeric_only=True)
```

```python
df_activ
```

```python
sns.set(rc={'figure.figsize':(13,6)}) # Augmenter la taille du graphique
labels = activités[1:]
men_means = list(df_activ.iloc[1,:])
women_means = list(df_activ.iloc[0,:])

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Moyennes')
ax.set_title("Moyenne des notes d'importance attribuées aux différentes activités")
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
