```python
def ecart_pref(index):
    s = 0
    for i in range(0,len(df_pref.columns)):
                   s = s + (df_pref[df_pref.columns[i]][index] - df_pref_o[df_pref_o.columns[i]][index])**2
    return( np.sqrt(s))
#Cette fonction renvoie la distance moyenne entre les prÃ©fÃ©rences et les prÃ©fÃ©rences du partenaires
```

```python
def note_moyenne(index):
    m = 1/100 * (df['attractive_important'][index]*df['attractive_partner'][index] + df['sincere_important'][index]*df['sincere_partner'][index] + df['intelligence_important'][index]*df['intelligence_partner'][index] + df['funny_important'][index]*df['funny_partner'][index] + df['ambition_important'][index]*df['ambition_partner'][index] + df['shared_interests_important'][index]*df['shared_interests_partner'][index])
    return(m)
#Cette fonction retourne la note moyenne attribuÃ©e par la personne qui note la personne en face (pondÃ©rÃ©e par l'importance qu'elle attribue aux 6 caractÃ©ristiques Ã©valuÃ©es)
```

```python
def note_moyenne_o(index):
    m = 1/100 * (df['pref_o_attractive'][index]*df['attractive_o'][index] + df['pref_o_sincere'][index]*df['sincere_o'][index] + df['pref_o_intelligence'][index]*df['intelligence_o'][index] + df['pref_o_funny'][index]*df['funny_o'][index] + df['pref_o_ambitious'][index]*df['ambitious_o'][index] + df['pref_o_shared_interests'][index]*df['shared_interests_o'][index])
    return(m)
#Cette fonction donne la note moyenne attribuÃ©e par le partenaire de la personne principale (elle fait la mÃªme chose que la fonction prÃ©cÃ©dente mais du point de vue du partenaire)
```

```python
def note_moyenne_propre(index):
    m = 1/100 * (df['attractive_important'][index]*df['attractive'][index] + df['sincere_important'][index]*df['sincere'][index] + df['intelligence_important'][index]*df['intelligence'][index] + df['funny_important'][index]*df['funny'][index] + df['ambition_important'][index]*df['ambition'][index])
    return(m)
#Note moyenne de soi perÃ§ue par soi 
```

```python
df_bis = df[['attractive' , 'sincere' , 'intelligence' , 'funny' , 'ambition' , 'attractive_o' , 'sincere_o' , 'intelligence_o' , 'funny_o' , 'ambitious_o']]
#Dans cette base de donnÃ©e, j'ai les colonnes des notes que l'individu se donne Ã  lui mÃªme (son propre avis sur lui) et ensuite j'ai les colonnes de ces mÃªmes caractÃ©ristiques mais notÃ©es par le partenaire (_o)
```

```python
def liste_biais_note(index):
    liste = []
    liste_colonnes = df_bis.columns
    for i in range(0 , (len(liste_colonnes))//2):
        liste.append(df_bis[liste_colonnes[i]][index] - df_bis[liste_colonnes[i+5]][index])
    return(liste)
  #Renvoie la liste des biais (i.e note perÃ§ue - note reÃ§ue par le partenaire)       
```

```python
def biais_moyen_anticipe(index):
    B = liste_biais_note(index)
    m = 0
    for i in range(0,len(B)):
        m = m + B[i]*( df_pref[df_pref.columns[i]][index] )
    return(m/100)
#Renvoie le biais moyen perÃ§u par soi mÃªme
```

```python
def biais_moyen_anticipe_o(index):
    B = liste_biais_note(index)
    m = 0
    for i in range(0,len(B)):
        m = m + B[i]*( df_pref_o[df_pref_o.columns[i]][index] )
    return(m/100)
#Renvoie le biais moyen perÃ§u par le partenaire
```

```python
def biais_moyen_reel(index):
    return(note_moyenne_propre(index) - note_moyenne_o(index)) 
```
